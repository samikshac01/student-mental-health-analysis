# 🧠 Student Mental Health Analysis

## 📊 Objective
The goal of this project is to analyze various factors affecting student mental health, such as academic pressure, sleep patterns, and financial concerns.

## 🛠️ Tools Used
- Python
- Pandas
- Matplotlib
- Seaborn

## 📁 Dataset
Student Mental Health Survey Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Setup successful ✅")
```

### Data Loading

```python
df = pd.read_csv("data/MentalHealthSurvey.csv")
df.head()
```

```python
print(df.columns)
```

### data cleaning

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.columns)
```

```python
df.info()
df.describe()
```

## 📊 Exploratory Data Analysis (EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

### Insight : Gender Distribution 
-The dataset contains a higher number of males compared to female students.
-This indicates an imbalance in gender representation.
-As a result , the analysis may be slightly biased towards mae student experience .

```python
sns.boxplot(x='depression', y='average_sleep', data=df)
plt.title("Sleep vs depression")
plt.show()
```

### 🔍 Insight: Sleep vs Depression

- Students who sleep fewer hours tend to show higher levels of depression.
- This highlights the importance of adequate sleep for maintaining good mental health.
- Ensuring 7–8 hours of sleep may help reduce depression levels among students.


```python
sns.boxplot(x='academic_pressure', y='anxiety', data=df)
plt.title("Academic Pressure vs Anxiety")
plt.show()
```

### 🔍 Insight: Academic Pressure vs Anxiety

- As academic pressure increases, the anxiety levels among students also increase.
- This indicates a strong relationship between academic workload and mental health.
- High academic pressure can negatively impact students' well-being, suggesting the need for better workload management.


```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correaltion Heatmap")
plt.show()
```

### 🔍 Insight: Correlation Analysis (Heatmap)

- The heatmap shows a positive correlation between academic pressure, anxiety, and depression.
- This means that as academic pressure increases, both anxiety and depression levels tend to rise.
- Sleep shows a negative correlation with depression, indicating that higher sleep duration is associated with lower depression levels.
- Financial concerns and academic workload also show moderate correlation with stress-related factors.
- Overall, multiple factors such as academic pressure, sleep, and financial issues collectively impact student mental health.

```python
sns.boxplot(x='financial_concerns', y='depression', data=df)
plt.title("Financial concerns vs depression")
plt.show()
```

### Insight- Financial concerns vs depression 
-student with higher financial concerns tends to show higher level of depression.
-Financial stress apprears to be a significant factor affecting student mental health.

```python
sns.boxplot(x='social_relationships', y='anxiety', data=df)
plt.title("Social Relationship Vs Anxiety")
plt.show()
```

### Insights - Social Relationship Vs Anxiety
-Student with weaker social relationship tend to experience higher anxiety level.
-Strong social support tends to appears to reduce mental stress.
-Encouraging social interaction my help improve mental health among students.

```python
df['average_sleep'].unique()
```

```python
def convert_sleep(x):
    parts = x.split('-')
    low = int(parts[0])
    high = int(parts[1].split()[0])
    return (low + high) / 2

df['sleep_numeric'] = df['average_sleep'].apply(convert_sleep)
```

```python
high_risk = df[(df['academic_pressure'] > 3) & (df['sleep_numeric'] < 6)]
high_risk.head()
```

### Insights : High_Risk Student group
-Students with high academic pressure , financial concerns , low sleep duration and low social interaction form a high risk group.
-These students show higher level of depression and anxiety.
-This suggests both academic pressure and poor sleep habits together significantly imapct mmental health.

## 🎯 Key Findings

- Academic pressure, sleep, and financial concerns are major factors affecting student mental health.
- Social relationships play an important role in reducing anxiety.
- A specific group of students (high pressure + low sleep) is more vulnerable.
- Mental health support should focus on lifestyle and academic balance.

## 📌 Final Conclusion

- This analysis highlights that student mental health is influenced by multiple factors such as academic pressure, sleep patterns, and financial concerns.
- Higher academic pressure is strongly associated with increased anxiety and stress levels.
- Students with lower sleep duration tend to experience higher levels of depression.
- Financial and academic challenges also contribute to mental health issues.
- Overall, maintaining a balanced academic workload and healthy lifestyle is essential for improving student well-being.
